**find all links in the given page**

```python
import requests
from bs4 import BeautifulSoup
import csv


url = 'https://www.w3schools.com/html/default.asp'
page = requests.get(url)


soup = BeautifulSoup(page.content, 'html.parser')
print(soup)
# find all the <a> tags
links = soup.find_all('a', href=True)

for link in links:
  print("https://www.w3schools.com"+str(link['href']))
```

```python
import requests
from bs4 import BeautifulSoup
import csv


url = 'https://www.w3schools.com/'
page = requests.get(url)

soup = BeautifulSoup(page.content, 'html.parser')

# find all the <a> tags
links = soup.find_all('a', href=True)

# create a CSV file
with open('links.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    for link in links:
        writer.writerow(["https://www.w3schools.com/"+link['href']])
```

**follow link and search for keyword**

```python
import requests
from bs4 import BeautifulSoup
import csv
import re

url = 'http://mgmcen.ac.in/'
page = requests.get(url)

soup = BeautifulSoup(page.content, 'html.parser')


links = soup.find_all('a', href=True)
dict1={}
wordtosearch="vision"
for link in links:
  a=link['href']
  if(a[0:5]!="http"):
    a="http://mgmcen.ac.in/"+a
  try:
    page= requests.get(a)
  except:
    continue
  if re.search(wordtosearch, page.text):
    dict1[a]="word "+wordtosearch+"  -found"
  else:
    dict1[a]="word "+wordtosearch+" -not found"
#output in csv
with open('wordsearch.csv', 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Link','Word Search'])
    for key,value in dict1.items():
      writer.writerow([key,value])
```

**find all links from flipkart products**

```python
import requests
from bs4 import BeautifulSoup
import csv

url = 'https://www.flipkart.com/search?q=mobile&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_2_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_2_na_na_na&as-pos=1&as-type=RECENT&suggestionId=mobile%7CMobiles&requestId=08e2bc91-2429-4f16-9573-4aa6d08e404e&as-backfill=on&p%5B%5D=facets.ram%255B%255D%3D4%2BGB'


response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')


listings = soup.find_all('div', {'class': '_2kHMtA'})


with open('output.csv', 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Product Name', 'Link'])
    for listing in listings:
        name = listing.find('div', {'class': '_4rR01T'}).text
        link = 'https://www.flipkart.com' + listing.find('a', {'class': '_1fQZEK'})['href']
        writer.writerow([name, link])
```
