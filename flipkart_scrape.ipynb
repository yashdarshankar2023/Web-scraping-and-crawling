```python
import csv
import requests
from bs4 import BeautifulSoup

url_template = 'https://www.flipkart.com/search?q=mobile&otracker=search&otracker1=search&marketplace=FLIPKART&page={}'

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}

product_names = []
product_prices = []
product_links = []
product_ratings = []
product_discounts = []

for page_num in range(1, 11):
    url = url_template.format(page_num)
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, 'html.parser')

    for product in soup.find_all('div', {'class': '_2kHMtA'}):
        product_name = product.find('div', {'class': '_4rR01T'})
        product_price = product.find('div', {'class': '_30jeq3 _1_WHN1'})
        product_link = product.find('a', {'class': '_1fQZEK'})
        product_rating = product.find('div', {'class': '_3LWZlK'})
        product_discount = product.find('div', {'class': '_3Ay6Sb'})

        if product_name is not None and product_price is not None and product_link is not None:
            product_names.append(product_name.text)
            product_prices.append(int(product_price.text.replace('₹', '').replace(',', '')))
            product_links.append(product_link['href'])

            if product_rating is not None:
                product_ratings.append(product_rating.text)
            else:
                product_ratings.append('No rating')

            if product_discount is not None:
                product_discount_str = product_discount.text.strip().replace('off', '').replace('%', '').strip()
                if product_discount_str:
                    product_discounts.append(int(product_discount_str))
                else:
                    product_discounts.append(0)
            else:
                product_discounts.append(0)

with open('flipkart_mobiles.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Product Name', 'Product Price', 'Product Rating', 'Product Link', 'Discount'])

    for i in range(len(product_names)):
        writer.writerow([product_names[i], product_prices[i], product_ratings[i], 'https://www.flipkart.com' + product_links[i], product_discounts[i]])

print('Data saved to flipkart_mobiles.csv')
```

```python
import csv
import requests
from bs4 import BeautifulSoup

url_template = 'https://www.flipkart.com/search?q=mobile&otracker=search&otracker1=search&marketplace=FLIPKART&page={}'

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}

product_data = []

for page_num in range(1, 11):
    url = url_template.format(page_num)
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, 'html.parser')

    for product in soup.find_all('div', {'class': '_2kHMtA'}):
        product_name = product.find('div', {'class': '_4rR01T'})
        product_price = product.find('div', {'class': '_30jeq3 _1_WHN1'})
        product_link = product.find('a', {'class': '_1fQZEK'})
        product_rating = product.find('div', {'class': '_3LWZlK'})
        product_discount = product.find('div', {'class': '_3Ay6Sb'})
        product_camera = product.find('ul', {'class': '_1xgFaf'})
        product_ram = product.find('ul', {'class': '_1xgFaf'})
        product_battery = product.find('ul', {'class': '_1xgFaf'})
        product_display = product.find('ul', {'class': '_1xgFaf'})

        if product_name is not None and product_price is not None and product_link is not None:
            product_name_str = product_name.text.strip()
            product_price_str = product_price.text.strip().replace(',', '')[1:]
            product_link_str = 'https://www.flipkart.com' + product_link['href']
            product_rating_str = product_rating.text.strip() if product_rating is not None else 'No rating'
            product_discount_str = product_discount.text.strip().replace(' off', '').replace('%', '') if product_discount is not None else '0'
            product_camera_str = product_camera.find_all('li')[0].text.strip() if product_camera is not None else 'N/A'
            product_ram_str = product_ram.find_all('li')[1].text.strip() if product_ram is not None else 'N/A'
            product_battery_str = product_battery.find_all('li')[3].text.strip() if product_battery is not None else 'N/A'
            product_display_str = product_display.find_all('li')[2].text.strip() if product_display is not None else 'N/A'

            product_data.append([product_name_str, product_price_str, product_link_str, product_rating_str, product_discount_str, product_camera_str, product_ram_str, product_battery_str, product_display_str])

with open('flipkart_mobiles.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Product Name', 'Price (INR)', 'Product Link', 'Rating', 'Discount', 'Camera', 'RAM', 'Battery', 'Display'])
    writer.writerows(product_data)


```
