```python
import requests
from bs4 import BeautifulSoup
import urllib.request
import os

url = 'https://www.google.com/search?q=apple+images+jpg&tbm=isch&ved=2ahUKEwiq5fra56D-AhXE53MBHYa4CfkQ2-cCegQIABAA&oq=apple+images+jpg&gs_lcp=CgNpbWcQAzIFCAAQgAQyBggAEAgQHjIGCAAQCBAeOgQIIxAnOgcIABANEIAEUN0DWOEPYM4QaABwAHgAgAGIAYgBjgSSAQMwLjSYAQCgAQGqAQtnd3Mtd2l6LWltZ8ABAQ&sclient=img&ei=vsk0ZKqOHcTPz7sPhvGmyA8&bih=544&biw=1280'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')


img_tags = soup.find_all('img')
img_urls = []
for img_tag in img_tags:
    img_url = img_tag['src']
    img_urls.append(img_url)


output_folder = 'apple_images'
os.makedirs(output_folder, exist_ok=True)
count = 0
i = 0
while count < 500 and i < len(img_urls):
    img_url = img_urls[i]
    img_filename = os.path.join(output_folder, f'image_{i+1}.jpg')
    try
        urllib.request.urlretrieve(img_url, img_filename)
        print(f"Downloaded: {img_url} -> {img_filename}")
        count += 1
    except Exception as e:
        print(f"Error downloading image {img_url}: {e}")
    i += 1

print("Downloaded a total of", count, "images.")
```

```python
import csv
import requests
from bs4 import BeautifulSoup

url = 'https://www.imdb.com/title/tt1375666/reviews'

response = requests.get(url)

soup = BeautifulSoup(response.content, 'html.parser')

review_containers = soup.find_all('div', class_='lister-item-content')

with open('reviews.csv', 'w', newline='', encoding='utf-8') as csvfile:

    writer = csv.writer(csvfile)

    writer.writerow(['Review Title', 'Rating', 'Review Text'])

    for container in review_containers:
        review_title = container.find('a', class_='title').text.strip()
        review_rating_element = container.find('span', class_='rating-other-user-rating')
        review_rating = review_rating_element.text.strip() if review_rating_element else 'N/A'
        review_text = container.find('div', class_='text').text.strip()


        writer.writerow([review_title, review_rating, review_text])

        print('Review Title:', review_title)
        print('Rating:', review_rating)
        print('Review Text:', review_text)
        print('---')
```

```python
import csv
import requests
from bs4 import BeautifulSoup


url = 'https://en.wikipedia.org/wiki/Web_scraping'


response = requests.get(url)

soup = BeautifulSoup(response.content, 'html.parser')

title = soup.find(id='firstHeading').text
intro = soup.find(class_='mw-parser-output').p.text

with open('scraped_data.csv', 'w', newline='', encoding='utf-8') as csvfile:

    writer = csv.writer(csvfile)

    writer.writerow(['Title', 'Introduction'])
    writer.writerow([title, intro])

    print('Title:', title)
    print('Introduction:', intro)
```
